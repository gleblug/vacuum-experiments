```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from cycler import cycler
import matplotlib.colors as mcolors
import os
import numpy as np

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = [6, 4]
plt.rcParams['font.size'] = 12
plt.rcParams['lines.markersize'] = 2.5
plt.rcParams['xtick.minor.visible'] = True
plt.rcParams['ytick.minor.visible'] = True
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = "#aaaaaa"
plt.rcParams['grid.alpha'] = 0.2
plt.rcParams['axes.grid.which'] = 'both'
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]
```

```python
without_probes = np.genfromtxt('comsol/110um_without_probes.txt', skip_header=8)
with_probes = np.genfromtxt('comsol/110um_with_probes.txt', skip_header=8)

plt.ylim(1000, 3000)
plt.plot(with_probes[:, 0], with_probes[:, 1], '-o', label='(1) наличие теплоотвода через контакты')
plt.plot(without_probes[:, 0], without_probes[:, 1]+15, '-', label='(2) отсутствие теплоотвода')
plt.xlabel('Расстояние вдоль нити (м)')
plt.ylabel('Температура (K)')
plt.legend()
plt.show()
```
